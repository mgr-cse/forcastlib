```python
import pandas as pd
from forcastlib.models.bayesnn import BayesianNN
from forcastlib.datasets.twitter import tweets_data

from forcastlib.graph.graph import plot_graph

from matplotlib import pyplot as plt

import torch
```

```python
# get data
X, Y = tweets_data()

X_train = torch.from_numpy(X[:-30]).float()
Y_train = torch.from_numpy(Y[:-30]).float()

X_test = torch.from_numpy(X[-30:]).float()
Y_test = torch.from_numpy(Y[-30:]).float()
```

```python
plt.plot(Y_train)
plt.show()
```

```python
model = BayesianNN(len(X_train[0]), 1, [])
model.fit(X_train, Y_train)
```

```python
mean, std = model.predict(X_test)
```

```python

plot_graph(Y_train, Y_test, mean, -std, std)
```
